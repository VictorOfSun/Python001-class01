```python
import pandas as pd
import numpy as np
```

```python
data = pd.DataFrame({
    'id':np.arange(1,21),
    'age':np.random.randint(10,70,20)
})
```

```python
# 1. SELECT * FROM data
data
```

```python
# 2. SELECT * FROM data LIMIT 10;
data.head(10)
```

```python
# 3. SELECT id FROM data;  //id 是 data 表的特定一列
data['id']
```

```python
# 4. SELECT COUNT(id) FROM data;
data['id'].count()
```

```python
# 5. SELECT * FROM data WHERE id<1000 AND age>30;
data[ (data['id']<1000) & (data['age']<30) ]
```

```python
# 6. SELECT id,COUNT(DISTINCT order_id) FROM table1 GROUP BY id;
data['order_id'] = np.random.randint(0,10,20)
data
```

```python
data.drop_duplicates('order_id')[['id','order_id']].groupby('id').agg({'order_id' : 'count'})
```

```python
# 7. SELECT * FROM table1 t1 INNER JOIN table2 t2 ON t1.id = t2.id;
data2 = pd.DataFrame({
    'id' : np.arange(1,11),
    'age' : np.random.randint(10,70,10),
    'order_id' : np.random.randint(0,10,10)
})
data2
```

```python
pd.merge(data, data2, on= 'id', how='inner')
```

```python
# 8. SELECT * FROM table1 UNION SELECT * FROM table2;
pd.concat([data, data2])
```
